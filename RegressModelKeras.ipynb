```python
import keras
from keras.models import Sequential
from keras.layers import Dense
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('https://cocl.us/concrete_data')# getting data
data.head()
data.describe()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error 
```

```python
data_cols = data.columns
predictors = data[data_cols[data_cols != 'Strength']]
target = data['Strength']
n_cols = predictors.shape[1]
#dividing training and testing data set
# 30% of data is for testing 
pred_train, pred_test, target_train, target_test = train_test_split(predictors, target, test_size = 0.3 , random_state = 4)
pred_train.shape
```

```python
print('Train set:', pred_train.shape, target_train.shape)
print('Test set:', pred_test.shape, target_test.shape)
```

```python
#defining regression model
def regression_model():
    model = Sequential()
    model.add(Dense(10, activation = 'relu', input_shape = (n_cols,)))#input layer
    model.add(Dense(10, activation = 'relu'))#hidden layer
    model.add(Dense(1))
    
    #compile model
    model.compile(optimizer = 'adam', loss= 'mean_squared_error')
    return model
```

```python
err_lst = [];
model = regression_model()
for x in range(50):    
    
    model.fit(pred_train, target_train,validation_data=(pred_test, target_test),epochs = 50, verbose = 2)
    scores = model.evaluate(pred_test, target_test, verbose = 0)
    predictions = model.predict(pred_test)
    error = mean_squared_error(target_test, predictions)
    err_lst.append(error)
```

```python
for er in err_lst:
    print(er)
```

```python
from scipy import stats
n_arry = np.array(err_lst)
print(type(n_arry))
sts = stats.describe(n_arry)
sts
```

```python
nn_std = np.std(n_arry)
```

```python
nn_mean = np.mean(n_arry)
```

```python
print("For Part A:")
print("Standard Deviation of root mean squared error: {} \n Mean of Root mean squared error: {}".format(nn_std, nn_mean))
```

Part B: 

```python
prediction_norm = (predictors - predictors.mean()) / predictors.std()
err_lst_B = []
pnorm_train, pnorm_test, target_train, target_test = train_test_split(prediction_norm, target, test_size = 0.3 , random_state = 4)
for x in range(50):    
    model = regression_model()
    model.fit(pnorm_train, target_train, validation_data=(pnorm_test, target_test),epochs = 50, verbose = 2)
    
    predictions = model.predict(pnorm_test)
    error = mean_squared_error(target_test, predictions)
    err_lst_B.append(error)
```

```python
for er in err_lst_B:
    print(er)
```

```python
nn_std = np.std(err_lst_B)
```

```python
nn_mean = np.mean(err_lst_B)
```

```python
print("For Part C:")
print("Standard Deviation of root mean squared error: {} \n Mean of Root mean squared error: {}".format(nn_std, nn_mean))
```

Part C:

```python
err_lst_C = []
for x in range(50):    
    model = regression_model()
    model.fit(pnorm_train, target_train, validation_data=(pnorm_test, target_test),epochs = 100, verbose = 2)
    
    predictions = model.predict(pnorm_test)
    error = mean_squared_error(target_test, predictions)
    err_lst_C.append(error)
```

```python
nn_std = np.std(err_lst_C)
nn_mean = np.mean(err_lst_C)
print("For Part C:")
print("Standard Deviation of root mean squared error: {} \n Mean of Root mean squared error: {}".format(nn_std, nn_mean))
```

```python
def regression_model_D():
    model = Sequential()
    model.add(Dense(10, activation = 'relu', input_shape = (n_cols,)))#input layer
    model.add(Dense(10, activation = 'relu'))# first hidden layer
    model.add(Dense(10, activation = 'relu'))# second hidden layer
    model.add(Dense(10, activation = 'relu'))# third hidden layer
    model.add(Dense(1))
    
    #compile model
    model.compile(optimizer = 'adam', loss= 'mean_squared_error')
    return model
```

```python
model = regression_model_D()
err_lst_D = []
```

```python
for x in range(50):    
    model.fit(pnorm_train, target_train, validation_data=(pnorm_test, target_test),epochs = 100, verbose = 2)    
    predictions = model.predict(pnorm_test)
    error = mean_squared_error(target_test, predictions)
    err_lst_D.append(error)
```

```python
nn_std = np.std(err_lst_D)
nn_mean = np.mean(err_lst_D)
print("For Part D:")
print("Standard Deviation of root mean squared error: {} \n Mean of Root mean squared error: {}".format(nn_std, nn_mean))
```
